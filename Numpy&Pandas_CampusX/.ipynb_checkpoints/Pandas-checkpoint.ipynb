#### Find out Virat Kohli scored most runs against which 3 teams 
Fetch all rows where batsman is v kohli, then group by based on team, then do sum on batsman runs.

```python
runs_by_vk=delivery[delivery['batsman']=='V Kohli']
```

```python
runs_by_vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
def run_scored(batsman_name):
    runs_by_batsman=delivery[delivery['batsman']==batsman_name]
    return runs_by_batsman.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
print(run_scored('V Kohli'))
print(run_scored('DA Warner'))
print(run_scored('G Gambhir'))
print(run_scored('MS Dhoni'))
```

#### isin() function

Q. Find the most destructive death over batsman in the history of IPL

death over==> last 5 overs
destructive==> on basis of strike rate==> strike rate = (number of runs/number of balls)/100 and also the batsman has played minm 200 balls in over 16-20 in history of IPL

```python
# find balls faced by each batsman
delivery.groupby('batsman')['batsman_runs'].count().sort_values(ascending = False)
```

```python
# But what no. of balls faced by each batsman in death overs
death_overs=delivery[delivery['over']>15]
```

```python
# so balls faced by each batsman in death overs will be:- 
death_overs.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(10)
```

```python
# Apply legit condition that batsman faced 200 balls so that one who's playing first match & hits 6 or 4 doesn't gets unfair advantage
death_overs_batsman=death_overs.groupby('batsman')['batsman_runs'].count()
legit_batsman = death_overs_batsman>200
batsman_list=death_overs_batsman[legit_batsman].index.tolist()
batsman_list
```

##### Calculate strike rate of these 43 batsman
for that will've to calculate runs scored & balls played by these 43 batsman
1st option:- create 43 mask for each batsman & apply OR condition put in delivery.
Better optimized alternative:- use isin() funcn, if there's various possible condition for mask.


```python
# give only those batsman which are in batsman list
final=death_overs[death_overs['batsman'].isin(batsman_list)]
# final
runs= final.groupby('batsman')['batsman_runs'].sum()
# runs
balls=final.groupby('batsman')['batsman_runs'].count()
# balls
```

```python
sr=((runs/balls)*100).sort_values(ascending=False)
sr
```

#### Pivot
It gives data in a very comprehensive manner, and helps in making valuable interpretation, which isn't visible in original data.
Pivot table takes main data cols, as indexes, values or columns.

```python
# Random dataframe creation
import random

# Define possible values
names = ["Nitish", "Anu", "Mukku", "Suri", "Rajiv", "Neha", "Amit", "Pooja", "Arjun", "Meera"]
genders = ["M", "F"]
cities = ["Kolkata", "Gurgaon", "Patna", "Mumbai", "Delhi", "Bangalore", "Pune", "Hyderabad", "Chennai"]
frequencies = ["Daily", "Weekly", "Monthly", "Once", "Never"]
items = ["Burger", "Sandwich", "Vada", "Pizza", "Paneer", "Pasta", "Noodles", "Samosa", "Salad", "Taco"]

# Generate 50 unique rows
data = {
    "Name": [random.choice(names) for _ in range(50)],
    "Gender": [random.choice(genders) for _ in range(50)],
    "City": [random.choice(cities) for _ in range(50)],
    "Frequency": [random.choice(frequencies) for _ in range(50)],
    "Item": [random.choice(items) for _ in range(50)],
    "Spends": np.random.randint(10, 100, size=50)  # Random spending amounts
}

# Create DataFrame
df = pd.DataFrame(data)

# Display first few rows
print(df.head())
```

```python
df.to_csv("food.csv", index=False)
```

```python
food=pd.read_csv('food.csv')
food.head()
```

```python
 food.pivot_table(index=['City','Gender'],columns=['Item','Frequency'],values='Spends',aggfunc='mean')
```

### Handling missing values
#### 1. dropna() method
used to drop missing values from data

#### Using Titanic dataset

```python
data= pd.read_csv('Titanic.csv')
```

```python
data.shape
```

```python
value_data=data.dropna()
```

```python
value_data.shape #i.e. only 183 rows had values in all cols present, rest in any col values was missing
```

##### how parameter
Drop only those rows, where value for all cols are missing

```python
new_data=data.dropna(how='all')
new_data.shape # no rows are there for which all vals are null
```

##### axis parameter
Setting axis=1, then it drops on basis of cols, by default it's 0 which means it drops on row basis

```python
v_data=data.dropna(axis=1) #drops all those cols, which've even 1 missing value, heere also we can use how param
v_data.shape 
```

##### subset parameter
Here, can pass 1 or multiple columns to find and drop missing values, no need to apply it on whole data

```python
data.dropna(subset=['Cabin']).shape
```

```python
data.dropna(subset=['Embarked']).shape
```

```python
data.dropna(subset=['Cabin','Embarked']).shape #so its performing OR or UNION operation
```
