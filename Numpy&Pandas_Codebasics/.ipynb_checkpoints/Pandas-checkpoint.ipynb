#### Merge Dataframes
It's similar to SQL Joins

```python
df1= pd.DataFrame({
    "city": ["Chicago", "New York", "Orlando", "Baltimore"],
    "temperature": [21,14,35,32]
})
df1
```

```python
df2= pd.DataFrame({
    "city": ["Chicago", "New York", "San Francisco"],
    "humidity": [65,68,72]
})
df2
```

```python
df3= pd.merge(df1,df2, on="city") #by default it performs inner join
df3
```

```python
# outer join
df3= pd.merge(df1,df2, on="city", how= "outer", indicator=True) #indicator tells from which df particular row is coming
df3
```

```python
# left join
dfl= pd.merge(df1,df2, on="city", how= "left")
print(dfl)
# right join
dfr= pd.merge(df1,df2, on="city", how= "right")
print(dfr)
```

```python
df1= pd.DataFrame({
    "city": ["Chicago", "New York", "Orlando", "Baltimore"],
    "temperature": [21,14,35,32],
    "humidity":[65,68,71,75]
})
df1
```

```python
df3= pd.merge(df1,df2,on='city', suffixes=('_left','_right')) # by default it takes x & y as suffixes
df3
```

#### Crosstab
By default, computes a frequency table of the factors unless an array of values and an aggregation function are passed.

```python
df= pd.read_excel('13_survey.xls')
df
```

```python
#applied directly on pandas and not df
pd.crosstab(df.Nationality, df.Handedness)
```

```python
pd.crosstab(df.Sex, df.Handedness, margins=True) #setting margins =True, gives total value as well
```

```python
pd.crosstab(df.Sex, [df.Handedness, df.Nationality], margins=True) #we can've multiple values at col level
```

```python
# same can be done at row leevel, i.e. multiple values can be passed at row level
pd.crosstab([df.Nationality, df.Sex], df.Handedness, margins=True)
```

```python
# finding % instead of value
pd.crosstab([df.Sex], [df.Handedness], normalize='index')
```

```python
# find average age of females who're left handed, & male who're right handed 
pd.crosstab([df.Sex], [df.Handedness],values=df.Age,aggfunc=np.average )
```

```python
# Creating own custom calendar
from pandas.tseries.holiday import AbstractHolidayCalendar, nearest_workday, Holiday
class abvBirthdayCalendar(AbstractHolidayCalendar):
    """
    US Federal Government Holiday Calendar based on rules specified by:
    https://www.opm.gov/policy-data-oversight/
    snow-dismissal-procedures/federal-holidays/
    """
    rules=[
        Holiday("Atal Bihari Vajpayee's birthday", month=4, day=15)
    ]

myc=CustomBusinessDay(calendar=abvBirthdayCalendar())
myc
```

```python
pd.date_range(start='4/1/2017', end='4/30/2017',freq=myc)
```

```python
# in US, if holiday falls on weekends, it is observed prev weekday or next weekday,so we pass observance
# Creating own custom calendar
from pandas.tseries.holiday import AbstractHolidayCalendar, nearest_workday, Holiday
class abvBirthdayCalendar(AbstractHolidayCalendar):
    """
    US Federal Government Holiday Calendar based on rules specified by:
    https://www.opm.gov/policy-data-oversight/
    snow-dismissal-procedures/federal-holidays/
    """
    rules=[
        Holiday("Atal Bihari Vajpayee's birthday", month=4, day=16, observance=nearest_workday) 
        # can choose; nearest_workday, next_monday_or_tuesday, previous_friday etc..
    ]

myc=CustomBusinessDay(calendar=abvBirthdayCalendar())
myc
```

```python
pd.date_range(start='4/1/2017', end='4/30/2017', freq=myc)
```

```python
# In Egyptian country, weekends are fri & sat, and sun is normal working day, so can pass weekmask & set weekdays accordingly
eb=CustomBusinessDay(weekmask='Sun Mon Tue Wed Thu')
pd.date_range(start='7/1/2017', end='7/29/2017', freq=eb)
```

```python
# can also provide list of holidays 
eb=CustomBusinessDay(weekmask='Sun Mon Tue Wed Thu', holidays=["2017-07-04"])
pd.date_range(start='7/1/2017', end='7/29/2017', freq=eb)
```

#### Time Series: to_datetime
Most common problem in data analysis is lack of uniformity in the structure of input data.
eg:- 5th January 2017 can be written in different formats as:- 
* 2017-01-05
* Jan 5, 2017
* 01/05/2017
* 2017.01.05
* 2017/01/05

```python
dates= ['2017-01-05', 'Jan 5, 2017', '01/05/2017', '2017.01.05', '2017/01/05', '20170105']
pd.to_datetime(dates,format='mixed')
```

```python
# it can also handle time
dates= ['2017-01-05 2:30:00 PM', 'Jan 5, 2017 14:30:00', '01/05/2017', '2017.01.05', '2017/01/05', '20170105']
pd.to_datetime(dates,format='mixed')
```

US:- mm/dd/yyyy
Europe:- dd/mm/yyyy

```python
# to handle europe dates, we've to supply arg, bcoz by default it considers mm/dd/yyyy format
pd.to_datetime('5/1/2017', dayfirst=True)
```

```python
# custom format, let's take '$' as delimeter instead of '/'
pd.to_datetime('5$1$2017', format='%d$%m$%Y')
```

```python
# on passing garbage string
dates= ['abc', '2017-01-05 2:30:00 PM', 'Jan 5, 2017 14:30:00', '01/05/2017', '2017.01.05', '2017/01/05', '20170105']
pd.to_datetime(dates, errors='ignore')
# but here conversion is not performed for any of the values
```

#### Time Series: Timezone Handling
Two types of datetime objects in python:- 
* Naive(No timezone awareness)
* Time zone aware datetime

```python
df=pd.read_csv('20_ts_msft.csv', header=1, index_col='Date Time', parse_dates=True)
df
```

```python
df.index
```

```python
# converting naive datetime index to time zone aware
df=df.tz_localize(tz='US/Eastern')
df.index
```

-4 means US/Eastern time zone is 4 hrs behind the UTC zone 

```python
# converting these to berlin tiimezone index
df=df.tz_convert(tz='Europe/Berlin')
df
```

```python
# gives list of all timezones
from pytz import all_timezones
all_timezones
```

```python
# converting to calcutta time zone
df=df.tz_convert(tz='Asia/Calcutta')
df
```

```python
# can also apply the same on index
df.index=df.index.tz_convert(tz='Asia/Calcutta')
df
```

```python
# using timezone in date_range function
rng=pd.date_range(start='1/1/2017', periods=10,freq='h', tz='Europe/London')
rng
```

```python
# alternate
rng=pd.date_range(start='1/1/2017', periods=10, freq='h', tz='dateutil/Europe/London')
rng
```

NOTE:- difference between pytz and dateutil is:- pytz has complete list of predefined tz while dateutil uses all avlbl tz from your operating system
